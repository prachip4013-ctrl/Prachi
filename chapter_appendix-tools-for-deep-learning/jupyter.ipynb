```python

# Taking two numbers
a = 15
b = 5
sum_result = a + b
print("sum =", sum_result)

# Declare variables
a = 2         # integer
b = 1.3       # float
c = "Hello"   # string

# Display variables
print("value of a :", a)
print("value of b :", b)
print("value of c :", c)

# Display types of variables
print("type of a :", type(a))
print("type of b :", type(b))
print("type of c :", type(c))

# Declare variables
a = 10        # integer
b = 6.7       # float

# Type casting
x = float(a)  # int to float
y = int(b)    # float to int

# Display result
print("original value of a :", a, "Type :", type(a))
print("After casting a to float :", x, "Type :", type(x))
print("After casting b to int :", y, "Type :", type(y))

# Declare variables
a = 10
b = 5

# Logical AND
print("a > 5 and b > 2 :", a > 5 and b > 2)

# Logical OR
print("a > 15 or b > 2 :", a > 15 or b > 2)
```
